```python
#1. Setup 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import IsolationForest

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("Libraries imported.")

#2. Load Dataset
# Update path if needed
base_path = r"C:\Users\ASUS\OneDrive\Documents\corizo\archive (2)"
train_file = base_path + r"\UNSW_NB15_training-set.csv"
test_file = base_path + r"\UNSW_NB15_testing-set.csv"

try:
    df_train = pd.read_csv(train_file)
    df_test = pd.read_csv(test_file)
    print(f"Train data loaded. Shape: {df_train.shape}")
    print(f"Test data loaded. Shape: {df_test.shape}")

    df = pd.concat([df_train, df_test], ignore_index=True)
    print(f"\nCombined shape: {df.shape}")
    print("\nCombined data head:")
    display(df.head())

except FileNotFoundError as e:
    print(f"Error: {e}. Dataset files not found.")
    import sys
    sys.exit("Exiting due to missing dataset files.")


#3. Data Preprocessing
orig_labels = df[['label', 'attack_cat']].copy()

df_proc = df.drop(columns=['id', 'label', 'attack_cat'], errors='ignore')

print("\nShape after dropping cols:", df_proc.shape)
print("\nMissing values (before fill):", df_proc.isnull().sum().sum())

for col in df_proc.columns:
    if df_proc[col].dtype == 'object':
        df_proc[col].fillna('unknown', inplace=True)
    else:
        df_proc[col].fillna(0, inplace=True)

print("Missing values (after fill):", df_proc.isnull().sum().sum())

num_feats = df_proc.select_dtypes(include=np.number).columns.tolist()
cat_feats = df_proc.select_dtypes(include='object').columns.tolist()

print(f"\nNumerical features: {len(num_feats)}")
print(f"Categorical features: {len(cat_feats)}")

print("\nEncoding categorical features...")
for col in cat_feats:
    le = LabelEncoder()
    df_proc[col] = le.fit_transform(df_proc[col])
    print(f"  - Encoded '{col}'.")

print("\nScaling numerical features...")
scaler = StandardScaler()
df_proc[num_feats] = scaler.fit_transform(df_proc[num_feats])

print("\nProcessed data head (scaled/encoded):")
display(df_proc.head())
print(f"\nFinal processed shape: {df_proc.shape}")


#4. Isolation Forest Model
model = IsolationForest(contamination=0.01, random_state=42, n_estimators=100, n_jobs=-1, verbose=1)

print("\nTraining model...")
model.fit(df_proc)
print("Model training complete.")

a_scores = model.decision_function(df_proc)
a_labels = model.predict(df_proc)

df_proc['anomaly_score'] = a_scores
df_proc['anomaly_label'] = a_labels

print("\nAnomaly detection results:")
print(df_proc['anomaly_label'].value_counts())


#5. Analyze & Visualize

# Anomaly score distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_proc['anomaly_score'], bins=50, kde=True, color='skyblue')
plt.title('Anomaly Score Distribution')
plt.xlabel('Anomaly Score')
plt.ylabel('Count')

scores_sorted = np.sort(df_proc['anomaly_score'].values)
thresh_idx = int(len(scores_sorted) * model.contamination)
actual_thresh = scores_sorted[thresh_idx]

plt.axvline(x=actual_thresh, color='red', linestyle='--', label=f'Threshold ({actual_thresh:.4f})')
plt.legend()
plt.grid(axis='y', alpha=0.7)
plt.show()

print("\nSample of detected anomalies (label -1):")
anomalies_df = df_proc[df_proc['anomaly_label'] == -1]
display(anomalies_df.head(10))

# Original attack type distribution in anomalies
if not anomalies_df.empty:
    anom_orig_labels = orig_labels.loc[anomalies_df.index]
    print("\nOriginal attack type distribution in anomalies (top 10):")
    anom_type_counts = anom_orig_labels.groupby(['label', 'attack_cat']).size().sort_values(ascending=False)
    display(anom_type_counts.head(10))

    plt.figure(figsize=(14, 7))
    plot_df = anom_orig_labels['attack_cat'].value_counts().head(10).reset_index()
    plot_df.columns = ['Attack Category', 'Count']

    sns.barplot(x='Attack Category', y='Count', data=plot_df, palette='viridis')
    plt.title('Top 10 Attack Categories Detected as Anomalies')
    plt.xlabel('Original Attack Category')
    plt.ylabel('Number of Anomalies Detected')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.7)
    plt.tight_layout()
    plt.show()

else:
    print("\nNo anomalies detected based on contamination parameter.")


# Visualize features for anomalies vs. normal
temp_df_unscaled = df_proc.copy()
temp_df_unscaled[num_feats] = scaler.inverse_transform(temp_df_unscaled[num_feats])

feat1 = 'sbytes'
feat2 = 'dbytes'

plt.figure(figsize=(12, 8))
sns.scatterplot(x=feat1, y=feat2, hue='anomaly_label',
                data=temp_df_unscaled, palette='coolwarm',
                s=20, alpha=0.7)
plt.title(f'Anomaly Detection: {feat1} vs {feat2}')
plt.xlabel(f'{feat1} (unscaled, log scale)')
plt.ylabel(f'{feat2} (unscaled, log scale)')
plt.xscale('symlog')
plt.yscale('symlog')
plt.grid(True, alpha=0.6)
plt.legend(title='Anomaly Label')
plt.show()

print("\nProject Complete")
print("Anomaly detection using Isolation Forest on UNSW-NB15 data is done.")
```
